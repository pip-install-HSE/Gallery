```python
def get_row(date_):
    in_t = datetime.datetime.fromtimestamp(date_['DateTimeInTick'] / 1000)
    out_t = datetime.datetime.fromtimestamp(date_['DateTimeOutTick'] / 1000)
    month_1 = in_t.month
    month_2 = out_t.month
    weekday_1 = in_t.weekday()
    weekday_2 = out_t.weekday()
    halfhour_1 = 2 * in_t.hour + 1 + int(in_t.minute >= 30)
    halfhour_2 = 2 * in_t.hour + 1 + int(in_t.minute >= 30)
    efficiency = date_['MacCountAll'] / date_['Duration']
    
    temp = pd.DataFrame()
    if month_1 != month_2 or weekday_1 != weekday_2 or halfhour_1 != halfhour_2:
        temp.append({'month': month_1, 'week_day': weekday_1, 'half_hour': halfhour_1, 'efficiency': efficiency}, ignore_index=True)
        temp.append({'month': month_2, 'week_day': weekday_2, 'half_hour': halfhour_2, 'efficiency': efficiency}, ignore_index=True)
        return temp
    else:
        temp.append({'month': month_1, 'week_day': weekday_1, 'half_hour': halfhour_1, 'efficiency': efficiency}, ignore_index=True)
        return temp
```

```python
new_df = pd.DataFrame()
new_df
```

```python
df.apply(lambda x: new_df.append(get_row(x), ignore_index=True), axis=1)
```
